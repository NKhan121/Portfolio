```python
categories = %sql select "CategoryID", "CategoryName", "Description" from categories;
categories
```

```python
#looking at the number of products sold through category_counts

category_counts = %%sql select "CategoryID", count(*) AS count\
                    from  products \
                    group by "CategoryID" \
                    order by count desc;
            
category_counts
```

```python
pd.merge(category_counts.DataFrame(), categories.DataFrame())
```

```python
#how many products are discontinued
%%sql
SELECT "CategoryName", COUNT("ProductName") FROM categoriesmerged WHERE "Discontinued" = 0 GROUP BY "CategoryName";


%%sql
select "CategoryID", count(*) AS count
from  products
where "Discontinued" = 0
group by "CategoryID"
order by count desc;
```

```python
#selecting the 5 most expensive products
%%sql
SELECT "ProductName", "UnitPrice" FROM products WHERE "Discontinued" = 0 ORDER By "UnitPrice" DESC limit 5;
```
