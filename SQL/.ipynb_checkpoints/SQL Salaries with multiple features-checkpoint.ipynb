```python
#Importing 

import pandas as pd
from pandas.io import sql
import sqlite3
import numpy as np

conn = sqlite3.connect('../../assets/datasets/adult.sqlite')
df = sql.read_sql("SELECT * FROM adult",con = conn)
df.head()
```

```python
#Seeing how many nulls in the df
#looks like workclass, occupation and native-
df.isnull().sum()
```

```python
#drop missing values
df_no_missing=df.dropna()
df_no_missing
```

```python
df_cleaned = df.dropna(how='all')
df_cleaned
```

```python
# looking for salary above $50K
df['target'] = np.where(df['salary'] == str('>50K'),0,1)
df.head()
```

```python
df.dtypes
```

```python
#distribution of age
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')


df.groupby('salary').age.plot(kind='hist', legend = True)
#separate the age from above and below $50k
```

The only thing that is clearly visible is that there are many more young people in the low income class.

```python
#creating mulitple bar charts to visualize relationships between different features and salary

df.groupby('salary').sex.value_counts().unstack(level=0).plot(kind='bar', legend=True)
plt.xlabel('sex')
plt.show()
df.groupby('salary').race.value_counts().unstack(level=0).plot(kind='bar', legend=True)
plt.xlabel('race')
plt.show()
df.groupby('salary').education.value_counts().unstack(level=0).plot(kind='bar', legend=True)
plt.xlabel('education')
plt.show()
df.groupby('salary').occupation.value_counts().unstack(level=0).plot(kind='bar', legend=True)
plt.xlabel('occupation')
plt.show()
df.groupby('salary').workclass.value_counts().unstack(level=0).plot(kind='bar', legend=True)
plt.xlabel('workclass')
plt.show()
df.groupby('salary').workclass.value_counts().unstack(level=0).plot(kind='bar', legend=True)
plt.xlabel('marital-status')
plt.show()
```
